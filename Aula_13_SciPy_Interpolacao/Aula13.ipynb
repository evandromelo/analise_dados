üìö Aula: Interpola√ß√£o Linear e Regress√£o Polinomial com SciPy  
üß† Objetivo:  
Aplicar interpola√ß√£o linear para estimar valores entre pontos conhecidos.  
Realizar regress√£o polinomial para modelar tend√™ncias em dados agr√≠colas.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
from scipy.optimize import curve_fit
```

üìà Exemplo 1: Interpola√ß√£o Linear  
Dados de umidade do solo (%) x Profundidade (cm)

```python
# Dados amostrais: profundidade (cm) e umidade (%)
profundidade = np.array([0, 10, 20, 30, 40, 50])
umidade = np.array([30, 28, 25, 20, 18, 15])

# Criando fun√ß√£o de interpola√ß√£o linear
f_interp = interp1d(profundidade, umidade, kind='linear')

# Profundidade para interpolar (ex: 15 cm)
nova_profundidade = 15
nova_umidade = f_interp(nova_profundidade)

print(f"Umidade estimada na profundidade de {nova_profundidade} cm: {nova_umidade:.2f}%")
```

üìâ Exemplo 2: Regress√£o Polinomial  
Dados de crescimento de plantas (altura em cm) x Dias ap√≥s plantio

```python
# Dados: dias ap√≥s plantio e altura da planta (cm)
dias = np.array([1, 3, 5, 7, 9, 11, 13, 15])
altura = np.array([2, 5, 9, 14, 18, 21, 23, 24])

# Definindo fun√ß√£o polinomial de grau 2 (quadr√°tica)
def modelo_pol2(x, a, b, c):
    return a * x**2 + b * x + c

# Ajustando o modelo aos dados
popt, _ = curve_fit(modelo_pol2, dias, altura)

# Extraindo coeficientes
a, b, c = popt
print(f"Modelo ajustado: altura = {a:.3f}*dias^2 + {b:.3f}*dias + {c:.3f}")

# Valores previstos pelo modelo
altura_prevista = modelo_pol2(dias, a, b, c)

# Visualiza√ß√£o dos dados e do modelo
plt.figure(figsize=(8, 6))
plt.scatter(dias, altura, label='Dados reais', color='blue')
plt.plot(dias, altura_prevista, label='Regress√£o polinomial (grau 2)', color='red')
plt.xlabel('Dias ap√≥s plantio')
plt.ylabel('Altura da planta (cm)')
plt.title('Crescimento de Plantas - Regress√£o Polinomial')
plt.legend()
plt.grid(True)
plt.show()
```

 Exerc√≠cio: Estimar altura da planta no dia 12

```python
dia_alvo = 12
altura_estimada = modelo_pol2(dia_alvo, a, b, c)
print(f"Altura estimada no dia {dia_alvo}: {altura_estimada:.2f} cm")
```

üìå **Conclus√£o:**

**Interpola√ß√£o linear** √© √∫til para preencher dados entre pontos conhecidos.

**Regress√£o polinomial** permite modelar tend√™ncias e prever comportamentos em sistemas agr√≠colas.

Ambas as t√©cnicas s√£o √∫teis para **an√°lise de dados agr√≠colas e tomada de decis√£o.**
