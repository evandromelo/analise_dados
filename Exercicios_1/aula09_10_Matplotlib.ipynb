Histograma *plt.hist()*


Visualizar a distribuição de uma única variável numérica contínua. Ajuda a entender a frequência de diferentes valores e identificar a forma da distribuição (normal, assimétrica, etc.).  


Distribuição do rendimento por talhão, distribuição da altura de plantas, frequência de leituras de pH do solo,
variação de peso de animais.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Dados simulados: Rendimento de grãos em 200 parcelas (kg/ha)
# Supondo uma distribuição com média de 3500 kg/ha e desvio padrão de 400
np.random.seed(42) # Para reprodutibilidade
rendimento_graos = np.random.normal(loc=3500, scale=400, size=200)
# Adicionar alguns outliers para ver o efeito
rendimento_graos = np.append(rendimento_graos, [2000, 5000, 1500]) 
rendimento_graos = np.clip(rendimento_graos, 0, None) # Não aceita rendimento negativo

# Criando a figura e os eixos
fig, ax = plt.subplots(figsize=(10, 6))

# Plotando o histograma
# bins: número de barras ou bordas das barras
# color: cor das barras
# edgecolor: cor da borda das barras
# alpha: transparência das barras
ax.hist(rendimento_graos, bins=20, color='olivedrab', edgecolor='black', alpha=0.8)

# Adicionando título e rótulos
ax.set_title('Distribuição de Rendimento de Grãos (kg/ha) em Talhões', fontsize=16)
ax.set_xlabel('Rendimento de Grãos (kg/ha)', fontsize=12)
ax.set_ylabel('Frequência de Talhões', fontsize=12)

# Adicionando uma linha para a média
media_rendimento = np.mean(rendimento_graos)
ax.axvline(media_rendimento, color='red', linestyle='dashed', linewidth=2, label=f'Média: {media_rendimento:.2f} kg/ha')

ax.legend()
ax.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Boxplot *plt.boxplot()*


Visualizar a distribuição de uma variável numérica através de seus quartis. Excelente para identificar mediana, dispersão, assimetria e outliers de forma concisa.


Comparação de dados de sensores (temperatura, pH), resultados de experimentos (altura de plantas sob diferentes tratamentos),  distribuição de pesos de animais, variação do teor de açúcar em frutas.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Dados simulados: Altura de plantas de milho (cm) para um único tratamento
np.random.seed(43)
altura_plantas = np.random.normal(loc=180, scale=15, size=100)
altura_plantas = np.append(altura_plantas, [120, 230, 240]) # Outliers
altura_plantas = np.clip(altura_plantas, 0, None)

# Criando a figura e os eixos
fig, ax = plt.subplots(figsize=(8, 7))

# Plotando o boxplot
# patch_artist=True: preenche as caixas com cor
# medianprops: propriedades da linha da mediana
# flierprops: propriedades dos outliers
boxplot = ax.boxplot(altura_plantas, patch_artist=True, vert=True,  # vert=True para vertical
                     medianprops=dict(color='red', linewidth=2),
                     flierprops=dict(marker='o', markersize=8, markerfacecolor='orange', alpha=0.7))

# Personalizando a cor da caixa
for patch in boxplot['boxes']:
    patch.set_facecolor('lightgreen')

# Adicionando título e rótulos
ax.set_title('Distribuição da Altura de Plantas de Milho (Tratamento Padrão)', fontsize=16)
ax.set_ylabel('Altura da Planta (cm)', fontsize=12)
ax.set_xticks([]) # Remove os rótulos do eixo X, já que é uma única categoria

ax.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Boxplots Múltiplos:  *plt.boxplot()* com lista de dados  


Comparar a distribuição de uma variável numérica entre diferentes grupos ou categorias. Ideal para experimentos com múltiplos tratamentos.  

Comparação do rendimento entre diferentes variedades de sementes,  efeitos de diferentes doses de fertilizante na produtividade,  análise de qualidade da água em diferentes pontos de coleta,  comparação de altura de plantas sob diferentes regimes de irrigação.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Dados simulados: Rendimento de milho (ton/ha) para 3 variedades
np.random.seed(44)
rendimento_variedade_A = np.random.normal(loc=8.0, scale=0.8, size=100) # Variedade A
rendimento_variedade_B = np.random.normal(loc=9.5, scale=0.7, size=100) # Variedade B
rendimento_variedade_C = np.random.normal(loc=7.5, scale=1.0, size=100) # Variedade C
rendimento_variedade_C = np.append(rendimento_variedade_C, [4.0, 11.0]) # Outliers para C

dados_rendimento = [rendimento_variedade_A, rendimento_variedade_B, rendimento_variedade_C]
rotulos = ['Variedade A', 'Variedade B', 'Variedade C']

# Criando a figura e os eixos
fig, ax = plt.subplots(figsize=(12, 7))

# Plotando os boxplots múltiplos
boxplot = ax.boxplot(dados_rendimento, patch_artist=True, vert=True, 
                     medianprops=dict(color='red', linewidth=2),
                     flierprops=dict(marker='o', markersize=8, markerfacecolor='purple', alpha=0.7),
                     labels=rotulos) # Definindo os rótulos do eixo X

# Personalizando as cores das caixas
cores = ['lightsteelblue', 'lightcoral', 'lightgreen']
for patch, color in zip(boxplot['boxes'], cores):
    patch.set_facecolor(color)

# Adicionando título e rótulos
ax.set_title('Comparação de Rendimento de Milho por Variedade (ton/ha)', fontsize=16)
ax.set_xlabel('Variedade de Milho', fontsize=12)
ax.set_ylabel('Rendimento (ton/ha)', fontsize=12)

ax.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Heatmap *plt.imshow()* ou *sns.heatmap()*

Visualizar dados em uma grade bidimensional onde os valores são representados por cores. Excelente para identificar padrões em matrizes, como correlações ou dados espaciais (mapas de campo).

 Mapas de produtividade de campo, matrizes de correlação entre variáveis ambientais e de cultura, distribuição espacial de pragas, mapas de pH ou nutrientes do solo, dados de temperatura de estufas em grade.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns # Seaborn facilita a criação de heatmaps

# Dados simulados: Matriz de produtividade (ton/ha) de um campo dividido em 10x10 seções
# Simulação de um padrão de produtividade com base na posição
np.random.seed(45)
produtividade_campo = np.zeros((10, 10))

for i in range(10):
    for j in range(10):
        # Produtividade maior no centro e diminuindo para as bordas
        produtividade_campo[i, j] = (20 - (abs(i-4.5) + abs(j-4.5))) * 0.5 + np.random.normal(0, 0.5)

produtividade_campo = np.clip(produtividade_campo, 0, None) # Não aceita produtividade negativa

# Criando a figura e os eixos
fig, ax = plt.subplots(figsize=(10, 8))

# Plotando o heatmap usando Seaborn (mais fácil para anotar e colorbar)
# cbar_kws: Personalização da colorbar
sns.heatmap(produtividade_campo, annot=True, fmt=".1f", cmap='YlGn', linewidths=.5, ax=ax,
            cbar_kws={'label': 'Produtividade (ton/ha)'})

# Adicionando título e rótulos
ax.set_title('Mapa de Produtividade de Campo (ton/ha)', fontsize=16)
ax.set_xlabel('Coluna da Seção', fontsize=12)
ax.set_ylabel('Linha da Seção', fontsize=12)

# Rotulando os ticks dos eixos para corresponder às seções (0-9)
ax.set_xticks(np.arange(0.5, 10.5, 1), labels=np.arange(0, 10, 1))
ax.set_yticks(np.arange(0.5, 10.5, 1), labels=np.arange(0, 10, 1))
ax.tick_params(left=True, bottom=True) # Garante que os ticks apareçam

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns # Para um heatmap mais bonito

# Dados simulados para o exemplo combinado
np.random.seed(46)
temperatura_solo_dia = np.random.normal(25, 3, 100)
rendimento_trat_a = np.random.normal(8, 0.7, 50)
rendimento_trat_b = np.random.normal(9, 0.5, 50)
dados_boxplot = [rendimento_trat_a, rendimento_trat_b]
labels_boxplot = ['Tratamento A', 'Tratamento B']

# Matriz de correlação simulada para Heatmap
correlacao_agricola = pd.DataFrame(np.random.rand(4, 4) * 2 - 1, 
                                   columns=['Umidade', 'Temp. Ar', 'Nitrogênio', 'Rendimento'],
                                   index=['Umidade', 'Temp. Ar', 'Nitrogênio', 'Rendimento'])
correlacao_agricola = correlacao_agricola.corr() # Simula uma matriz de correlação real
correlacao_agricola.iloc[np.tril_indices_from(correlacao_agricola, k=-1)] = np.nan # Remove duplicatas

# Criando uma figura com 2x2 subplots
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
fig.suptitle('Análise Avançada de Dados Agrícolas', fontsize=20, weight='bold')

# --- Subplot 1: Histograma ---
axes[0, 0].hist(temperatura_solo_dia, bins=15, color='salmon', edgecolor='black', alpha=0.7)
axes[0, 0].set_title('Distribuição da Temperatura do Solo (°C)', fontsize=14)
axes[0, 0].set_xlabel('Temperatura (°C)', fontsize=10)
axes[0, 0].set_ylabel('Frequência', fontsize=10)
axes[0, 0].grid(axis='y', linestyle='--', alpha=0.6)

# --- Subplot 2: Boxplots Múltiplos ---
boxplot2 = axes[0, 1].boxplot(dados_boxplot, patch_artist=True, labels=labels_boxplot, vert=True,
                             medianprops=dict(color='darkred', linewidth=2))
for patch, color in zip(boxplot2['boxes'], ['lightskyblue', 'lightgreen']):
    patch.set_facecolor(color)
axes[0, 1].set_title('Comparação de Rendimento (Tratamentos)', fontsize=14)
axes[0, 1].set_xlabel('Tratamento', fontsize=10)
axes[0, 1].set_ylabel('Rendimento (ton/ha)', fontsize=10)
axes[0, 1].grid(axis='y', linestyle='--', alpha=0.6)

# --- Subplot 3: Heatmap (Correlação) ---
# Usando o DataFrame de correlação
sns.heatmap(correlacao_agricola, annot=True, fmt=".2f", cmap='coolwarm', ax=axes[1, 0],
            linewidths=.5, linecolor='gray', cbar_kws={'label': 'Coeficiente de Correlação'})
axes[1, 0].set_title('Matriz de Correlação de Variáveis Agrícolas', fontsize=14)
axes[1, 0].tick_params(axis='x', rotation=45) # Rotaciona os rótulos do eixo X
axes[1, 0].set_xlabel('') # Remove rótulo duplicado se a matriz for simétrica
axes[1, 0].set_ylabel('')

# --- Subplot 4: Exemplo de Gráfico de Linha (Apenas para ilustrar outro tipo) ---
dias = np.arange(1, 11)
crescimento = np.array([0.5, 1.2, 2.5, 4.0, 6.0, 8.5, 11.0, 13.5, 15.0, 16.0]) + np.random.normal(0, 0.5, 10)
axes[1, 1].plot(dias, crescimento, marker='^', linestyle='-', color='darkcyan', label='Crescimento da Planta')
axes[1, 1].set_title('Crescimento Médio da Planta ao Longo dos Dias', fontsize=14)
axes[1, 1].set_xlabel('Dias Após Plantio', fontsize=10)
axes[1, 1].set_ylabel('Altura (cm)', fontsize=10)
axes[1, 1].legend()
axes[1, 1].grid(True, linestyle='--', alpha=0.6)

plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Ajusta o layout para evitar sobreposição, deixando espaço para o suptitle
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(43)

# Dados de exemplo: Consumo de água por dia para 3 sistemas de irrigação
dias = np.arange(1, 8)
consumo_irrigacao_gotejamento = 10 + 2 * dias + np.random.normal(0, 1, 7)
consumo_irrigacao_aspersao = 8 + 3 * dias + np.random.normal(0, 1.5, 7)
consumo_irrigacao_pivot = 12 + 1.5 * dias + np.random.normal(0, 0.8, 7)

fig, ax = plt.subplots(figsize=(12, 7))

# Linha 1: Gotejamento - linha sólida, marcadores circulares
ax.plot(dias, consumo_irrigacao_gotejamento, 
        color='blue', linestyle='-', marker='o', 
        linewidth=2, markersize=8, label='Gotejamento')

# Linha 2: Aspersão - linha tracejada, marcadores quadrados
ax.plot(dias, consumo_irrigacao_aspersao, 
        color='green', linestyle='--', marker='s', 
        linewidth=2, markersize=8, label='Aspersão')

# Linha 3: Pivô Central - linha pontilhada, marcadores em triângulo
ax.plot(dias, consumo_irrigacao_pivot, 
        color='purple', linestyle=':', marker='^', 
        linewidth=2, markersize=8, label='Pivô Central')

ax.set_title('Consumo de Água por Sistema de Irrigação ao Longo da Semana', fontsize=16)
ax.set_xlabel('Dia da Semana', fontsize=12)
ax.set_ylabel('Consumo de Água (mm/dia)', fontsize=12)
ax.legend(fontsize=10)
ax.grid(True, linestyle='--', alpha=0.7)

plt.show()
```

Tornar o gráfico autoexplicativo e profissional. Essencial para a comunicação de resultados. Relatórios técnicos, artigos científicos, apresentações de resultados de pesquisa.

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(44)

# Dados de exemplo: Rendimento sob dois tipos de fertilizante
doses_fertilizante = np.arange(0, 100, 10)
rendimento_organico = 0.8 * doses_fertilizante + np.random.normal(0, 5, 10) + 50
rendimento_quimico = 1.2 * doses_fertilizante + np.random.normal(0, 7, 10) + 40

fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(doses_fertilizante, rendimento_organico, 
        color='brown', linestyle='-', marker='o', label='Fertilizante Orgânico')
ax.plot(doses_fertilizante, rendimento_quimico, 
        color='teal', linestyle='--', marker='x', label='Fertilizante Químico')

# Customização do título
ax.set_title('Impacto de Diferentes Fertilizantes no Rendimento da Cultura', 
             fontsize=18, color='darkgreen', weight='bold')

# Customização dos rótulos dos eixos
ax.set_xlabel('Dose de Fertilizante (kg/ha)', 
              fontsize=14, color='darkslategray', fontname='Arial')
ax.set_ylabel('Rendimento da Cultura (ton/ha)', 
              fontsize=14, color='darkslategray', fontname='Arial')

# Customização da legenda
ax.legend(loc='upper left', # Posição da legenda
          fontsize=12, 
          frameon=True, # Adiciona moldura
          shadow=True, # Adiciona sombra
          facecolor='lightgray', # Cor de fundo
          edgecolor='darkgray') # Cor da borda

ax.grid(True, linestyle=':', alpha=0.6)
plt.show()
```

**Configuração dos Eixos (set_xlim, set_ylim, set_xticks, set_yticks, tick_params)**  
Controlar a escala, os marcadores e os rótulos dos eixos para focar em regiões de interesse ou melhorar a legibilidade. Destacar períodos críticos de crescimento, analisar intervalos específicos de parâmetros.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

np.random.seed(45)

# Dados de exemplo: Variação de pH do solo ao longo de 12 meses
meses = np.arange(1, 13)
ph_solo = 6.5 + 0.5 * np.sin(meses * np.pi / 6) + np.random.normal(0, 0.1, 12)
ph_solo = np.clip(ph_solo, 5.5, 7.5) # Limita os valores de pH a um intervalo razoável

fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(meses, ph_solo, color='saddlebrown', linestyle='-', marker='D', markersize=7, label='pH do Solo')

# Definir limites dos eixos
ax.set_xlim(0.5, 12.5) # Limites para os meses
ax.set_ylim(5.0, 8.0)  # Limites para o pH

# Definir os ticks (marcadores) no eixo X e seus rótulos
meses_nomes = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 
               'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
ax.set_xticks(meses)
ax.set_xticklabels(meses_nomes, rotation=45, ha='right', fontsize=10) # Rotação e alinhamento

# Definir os ticks no eixo Y (com espaçamento específico)
ax.set_yticks(np.arange(5.0, 8.1, 0.5))

# Customizar os parâmetros dos ticks
ax.tick_params(axis='both', # Afeta ambos os eixos
               length=7,    # Comprimento do tick
               width=2,     # Largura do tick
               colors='gray', # Cor dos ticks e rótulos
               labelsize=10) # Tamanho da fonte dos rótulos

ax.set_title('Variação Mensal do pH do Solo', fontsize=16)
ax.set_xlabel('Mês', fontsize=12)
ax.set_ylabel('pH do Solo', fontsize=12)
ax.legend()
ax.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout() # Ajusta o layout para evitar sobreposição dos rótulos
plt.show()
```

**Anotações e Texto (ax.annotate, ax.text)**  
Adicionar informações específicas a pontos de dados ou regiões do gráfico, destacando insights importantes.
Apontar eventos (aplicação de fertilizante, chuva forte), destacar outliers de produtividade, adicionar conclusões rápidas.

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(46)

# Dados de exemplo: Crescimento da cultura com um evento específico
dias = np.arange(0, 15)
crescimento = 0.5 * dias**2 + np.random.normal(0, 5, 15) + 5
crescimento = np.clip(crescimento, 0, None)

fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(dias, crescimento, color='green', linestyle='-', marker='o', label='Crescimento da Cultura')

# Anotação: Apontar um ponto específico com uma seta
evento_dia = 7
evento_valor = crescimento[evento_dia]
ax.annotate('Aplicação de Nutrientes', 
            xy=(evento_dia, evento_valor), # Ponto a ser anotado
            xytext=(evento_dia + 2, evento_valor + 20), # Posição do texto
            arrowprops=dict(facecolor='black', shrink=0.05), # Propriedades da seta
            fontsize=12, color='red', weight='bold',
            bbox=dict(boxstyle="round,pad=0.5", fc="yellow", ec="k", lw=1, alpha=0.7)) # Caixa de texto

# Texto livre: Adicionar uma observação geral
ax.text(0.5, 0.95, 'Foco no período de crescimento acelerado', 
        transform=ax.transAxes, # Coordenadas relativas ao Axes (0-1)
        fontsize=10, color='darkgray', ha='left', va='top', 
        bbox=dict(boxstyle="square,pad=0.3", fc="white", ec="darkgray", lw=0.8, alpha=0.6))

ax.set_title('Crescimento da Cultura e Eventos Chave', fontsize=16)
ax.set_xlabel('Dias Após Plantio', fontsize=12)
ax.set_ylabel('Altura da Planta (cm)', fontsize=12)
ax.legend()
ax.grid(True, linestyle='--', alpha=0.7)

plt.show()
```

**Estilos Globais e Temas (plt.style.use)**  
Aplicar um conjunto de configurações visuais predefinidas para dar um estilo consistente e profissional aos gráficos com poucas linhas de código. Manter a padronização visual em relatórios e apresentações da empresa/instituição.

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(47)

# Dados de exemplo: Flutuação de preços de um produto agrícola
semanas = np.arange(1, 21)
preco_produto = 50 + 5 * np.sin(semanas / 3) + np.random.normal(0, 2, 20)

# Opção 1: Sem estilo (para comparação)
fig1, ax1 = plt.subplots(figsize=(10, 5))
ax1.plot(semanas, preco_produto, label='Preço do Produto')
ax1.set_title('Preço do Produto Agrícola (Padrão Matplotlib)')
ax1.set_xlabel('Semana')
ax1.set_ylabel('Preço (R$)')
ax1.legend()
ax1.grid(True)
plt.show()

# Opção 2: Usando um estilo predefinido ('ggplot')
plt.style.use('ggplot') # Aplica o estilo ggplot

fig2, ax2 = plt.subplots(figsize=(10, 5))
ax2.plot(semanas, preco_produto, color='darkgreen', linestyle='-', marker='o', label='Preço do Produto')
ax2.set_title('Preço do Produto Agrícola (Estilo ggplot)')
ax2.set_xlabel('Semana')
ax2.set_ylabel('Preço (R$)')
ax2.legend()
plt.show()

# Opção 3: Usando outro estilo ('seaborn-v0_8-darkgrid')
plt.style.use('seaborn-v0_8-darkgrid') # Aplica o estilo seaborn

fig3, ax3 = plt.subplots(figsize=(10, 5))
ax3.plot(semanas, preco_produto, color='goldenrod', linestyle='--', marker='^', label='Preço do Produto')
ax3.set_title('Preço do Produto Agrícola (Estilo seaborn-darkgrid)')
ax3.set_xlabel('Semana')
ax3.set_ylabel('Preço (R$)')
ax3.legend()
plt.show()

# Você pode listar os estilos disponíveis com:
# print(plt.style.available)

# Importante: para reverter ao estilo padrão, use:
plt.style.use('default') # Volta ao estilo padrão do Matplotlib
```
