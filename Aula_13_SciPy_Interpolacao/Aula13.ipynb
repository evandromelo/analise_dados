ðŸ§  Objetivo:  
Aplicar interpolaÃ§Ã£o linear para estimar valores entre pontos conhecidos.  
Realizar regressÃ£o polinomial para modelar tendÃªncias em dados agrÃ­colas.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
from scipy.optimize import curve_fit
```

ðŸ“ˆ Exemplo 1: InterpolaÃ§Ã£o Linear  
Dados de umidade do solo (%) x Profundidade (cm)

```python
# Dados amostrais: profundidade (cm) e umidade (%)
profundidade = np.array([0, 10, 20, 30, 40, 50])
umidade = np.array([30, 28, 25, 20, 18, 15])

# Criando funÃ§Ã£o de interpolaÃ§Ã£o linear
f_interp = interp1d(profundidade, umidade, kind='linear')

# Profundidade para interpolar (ex: 15 cm)
nova_profundidade = 15
nova_umidade = f_interp(nova_profundidade)

print(f"Umidade estimada na profundidade de {nova_profundidade} cm: {nova_umidade:.2f}%")
```

ðŸ“‰ Exemplo 2: RegressÃ£o Polinomial  
Dados de crescimento de plantas (altura em cm) x Dias apÃ³s plantio

```python
# Dados: dias apÃ³s plantio e altura da planta (cm)
dias = np.array([1, 3, 5, 7, 9, 11, 13, 15])
altura = np.array([2, 5, 9, 14, 18, 21, 23, 24])

# Definindo funÃ§Ã£o polinomial de grau 2 (quadrÃ¡tica)
def modelo_pol2(x, a, b, c):
    return a * x**2 + b * x + c

# Ajustando o modelo aos dados
popt, _ = curve_fit(modelo_pol2, dias, altura)

# Extraindo coeficientes
a, b, c = popt
print(f"Modelo ajustado: altura = {a:.3f}*dias^2 + {b:.3f}*dias + {c:.3f}")

# Valores previstos pelo modelo
altura_prevista = modelo_pol2(dias, a, b, c)

# VisualizaÃ§Ã£o dos dados e do modelo
plt.figure(figsize=(8, 6))
plt.scatter(dias, altura, label='Dados reais', color='blue')
plt.plot(dias, altura_prevista, label='RegressÃ£o polinomial (grau 2)', color='red')
plt.xlabel('Dias apÃ³s plantio')
plt.ylabel('Altura da planta (cm)')
plt.title('Crescimento de Plantas - RegressÃ£o Polinomial')
plt.legend()
plt.grid(True)
plt.show()
```

 ExercÃ­cio: Estimar altura da planta no dia 12

```python
dia_alvo = 12
altura_estimada = modelo_pol2(dia_alvo, a, b, c)
print(f"Altura estimada no dia {dia_alvo}: {altura_estimada:.2f} cm")
```
