```python
# Visualizando Dados de Crescimento Vegetal

import matplotlib.pyplot as plt
horas = [1,2,3,4,5]
umidade = [25.0,22.5,20.1,18.0,16.2]
plt.plot(horas, umidade, marker='o')
plt.xlabel("Hora")
plt.ylabel("Teor de água (% b.s.)")
plt.title("Curva de Secagem")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.plot([1, 2, 3, 4], [10, 20, 25, 30])  # Dados fictícios de crescimento
plt.xlabel("Tempo (semanas)")  
plt.ylabel("Altura da planta (cm)")  
plt.show()
```

```python
import numpy as np
dias = np.arange(0, 30)
altura_tratamento1 = 0.5 * dias + np.random.normal(0, 1, 30)
altura_tratamento2 = 0.7 * dias + np.random.normal(0, 1, 30)

fig, ax = plt.subplots()
ax.plot(dias, altura_tratamento1, label="Baixa adubação")
ax.plot(dias, altura_tratamento2, label="Alta adubação")
ax.set_xlabel("Dias após plantio")
ax.set_ylabel("Altura (cm)")
ax.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Dados simulados de crescimento (cm) ao longo de 30 dias
dias = np.arange(1, 31)
tratamento_A = np.random.normal(0.5, 0.1, 30).cumsum()
tratamento_B = np.random.normal(0.7, 0.1, 30).cumsum()

fig, ax = plt.subplots()
ax.plot(dias, tratamento_A, label='Tratamento A (50% fertilizante)')
ax.plot(dias, tratamento_B, label='Tratamento B (100% fertilizante)')

ax.set_xlabel('Dias após plantio')
ax.set_ylabel('Altura da planta (cm)')
ax.set_title('Crescimento de milho sob diferentes tratamentos')
ax.legend()
plt.show()
```
